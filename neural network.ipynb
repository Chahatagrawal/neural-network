```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
```

```python
def create_neural_network():
    model = Sequential()
    model.add(Dense(units=8, input_dim=4, activation='relu'))   # Input layer with 4 input neurons and 8 neurons in the hidden layer
    model.add(Dense(units=1, activation='sigmoid'))             # Output layer with 1 neuron and sigmoid activation function
    return model
```

```python
def compile_model(model):
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
def generate_data():
    X_train = np.random.rand(100, 4)  # Generate 100 samples with 4 features each
    y_train = np.random.randint(2, size=(100, 1))  # Generate 100 random binary labels (0 or 1)
    return X_train, y_train

def train_model(model, X_train, y_train):
    model.fit(X_train, y_train, epochs=50, batch_size=32)
```

```python
def main():
    model = create_neural_network()
    compile_model(model)
    X_train, y_train = generate_data()
    train_model(model, X_train, y_train)

if __name__ == "__main__":
    main()
```
